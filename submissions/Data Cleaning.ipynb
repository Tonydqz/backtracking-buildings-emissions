```python
def convertHeightToMetric(string):
    #first identifying any entry with ","
    if "," in string:
        #getting rid of the comma, doesn't work since it turns the string into a list
        string = string.split(',')
        #if the entry contains both foot and inches
        if "\"" in string:
            #split them into just foot and inches 
            foot, inches = string.split("'")
            #deleting the last quotation mark
            inches = inches[:-1]
            #conversion math
            return round((int(foot)*12 + int(inches)) * 2.54, 3)
        #if the entry contains only foot
        else:
            #deleting the quotation mark
            foot = string[:-1]
            return round(int(foot)*12*2.54, 3)
    #problems: there are entry where there are three sets of number
    #ie. 60',50',30'
    #some of the inches are represented by a single quote instead of double quote, the current
    #code rely on double quote as identification for inches
mergedDataSet['Height'] = mergedDataSet['Height'].apply(convertHeightToMetric)
mergedDataSet
```
